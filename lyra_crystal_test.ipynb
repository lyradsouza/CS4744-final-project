````python
import hfst_dev as hfst
import graphviz
import random
````

Code for random samples

````python
def sample_input(x,n=8,cycles=3):
        x2 = x.copy()
        x2.input_project()
        x2.minimize()
        return(random.sample(set(x2.extract_paths(max_cycles=3).keys()),n))
def sample_output(x,n=8,cycles=3):
        x2 = x.copy()
        x2.output_project()
        x2.minimize()
        return(random.sample(set(x2.extract_paths(max_cycles=3).keys()),n))
````

#### CODE TO GET WORDS FROM ENGLISH ###

````python
def get_words(x):
        x2 = x.copy()
        x2.input_project()
        x2.minimize()
        return(set(x2.extract_paths().keys()))
````

Read the lexicon machine

````python
istream = hfst.HfstInputStream('English')
````

````python
istream.is_good()
````

````python
English = istream.read()
istream.close()
````

#### EXTRACTING SET OF WORDS

````python
strings = get_words(English)
````

````python
len(strings)
````

````python
strings_ENGLISH = sorted(strings)
````

````python
strings_CLEANED = list(map(lambda x: x.replace('|', ''), strings_ENGLISH))
````

````python
strings_MERGED = zip(strings_ENGLISH, strings_CLEANED)
````

````python
# Demo of strings MERGED

for item, (original, cleaned) in enumerate(zip(strings_ENGLISH, strings_CLEANED)):
    print(item, (original, cleaned))
````

## Examples similar to ps3av2

#### Example 1
Find the set of words that contain *dog* on the orthographic side.  The answer is the machine *dog*.

Use definitional style.

````python
defs = {'English':English}
````

````python
expr = 'English.u & ${dog}'
dog = hfst.regex(expr, definitions=defs)
sample_input(dog)
````

Note: it is okay to leave the vertical bars.

#### Example 2
Find the set of orthographic words, the phonetic form
of which contains the phonetic form of *plan*.

````python
expr = '[English .o. $[{plan} .o. English].l].u'
e2 = hfst.regex(expr, definitions=defs)
````

````python
sample_input(e2)
````
